```python
#   Google Drive
from google.colab import drive
drive.mount('/content/gdrive')

#   
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering, KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score

#  
file_path = "/content/gdrive/My Drive/working_memory/child_adolesc_outliers.csv"
data = pd.read_csv(file_path)

#     :  (0)   (1)
children_data = data[data['(0)/(1)'] == 0]
adolescents_data = data[data['(0)/(1)'] == 1]

#    
X_children = children_data[['WMem:  ,  1',
                            'WMem:  ,  2',
                            'WMem:  ,  3']].values

X_adolescents = adolescents_data[['WMem:  ,  1',
                                  'WMem:  ,  2',
                                  'WMem:  ,  3']].values

#    
plt.figure(figsize=(10, 7))
plt.title("  ")
dend = shc.dendrogram(shc.linkage(X_children, method='ward'))
plt.show()

#    
plt.figure(figsize=(10, 7))
plt.title("  ")
dend = shc.dendrogram(shc.linkage(X_adolescents, method='ward'))
plt.show()

#        
inertia_children = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_children)
    inertia_children.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), inertia_children, marker='o')
plt.title("   ")
plt.xlabel(" ")
plt.ylabel("")
plt.show()

#        
silhouette_scores_children = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_children)
    score = silhouette_score(X_children, kmeans.labels_)
    silhouette_scores_children.append(score)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), silhouette_scores_children, marker='o')
plt.title("   ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()

#     
optimal_clusters_children = 3

#        
inertia_adolescents = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_adolescents)
    inertia_adolescents.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), inertia_adolescents, marker='o')
plt.title("   ")
plt.xlabel(" ")
plt.ylabel("")
plt.show()

#        
silhouette_scores_adolescents = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_adolescents)
    score = silhouette_score(X_adolescents, kmeans.labels_)
    silhouette_scores_adolescents.append(score)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), silhouette_scores_adolescents, marker='o')
plt.title("   ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()

#     
optimal_clusters_adolescents = 3

#    
agglo_children = AgglomerativeClustering(n_clusters=optimal_clusters_children, affinity='euclidean', linkage='complete')
children_clusters = agglo_children.fit_predict(X_children)

#    
agglo_adolescents = AgglomerativeClustering(n_clusters=optimal_clusters_adolescents, affinity='euclidean', linkage='complete')
adolescents_clusters = agglo_adolescents.fit_predict(X_adolescents)

#      3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_children[:, 0], X_children[:, 1], X_children[:, 2], c=children_clusters, cmap='viridis')
ax.set_title("  ")
ax.set_xlabel(' 1')
ax.set_ylabel(' 2')
ax.set_zlabel(' 3')
plt.show()

#      3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_adolescents[:, 0], X_adolescents[:, 1], X_adolescents[:, 2], c=adolescents_clusters, cmap='viridis')
ax.set_title("  ")
ax.set_xlabel(' 1')
ax.set_ylabel(' 2')
ax.set_zlabel(' 3')
plt.show()
```

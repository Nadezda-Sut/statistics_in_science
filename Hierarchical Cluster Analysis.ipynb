```python
# Подключение к Google Drive
from google.colab import drive
drive.mount('/content/gdrive')

# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering, KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score

# Загрузка данных
file_path = "/content/gdrive/My Drive/working_memory/child_adolesc_outliers.csv"
data = pd.read_csv(file_path)

# Разделение данных на две группы: дети (0) и подростки (1)
children_data = data[data['Ребенок(0)/Подросток(1)'] == 0]
adolescents_data = data[data['Ребенок(0)/Подросток(1)'] == 1]

# Извлечение данных для анализа
X_children = children_data[['WMem: Количество объектов, серия 1',
                            'WMem: Количество объектов, серия 2',
                            'WMem: Количество объектов, серия 3']].values

X_adolescents = adolescents_data[['WMem: Количество объектов, серия 1',
                                  'WMem: Количество объектов, серия 2',
                                  'WMem: Количество объектов, серия 3']].values

# Построение дендрограммы для детей
plt.figure(figsize=(10, 7))
plt.title("Дендрограмма для детей")
dend = shc.dendrogram(shc.linkage(X_children, method='ward'))
plt.show()

# Построение дендрограммы для подростков
plt.figure(figsize=(10, 7))
plt.title("Дендрограмма для подростков")
dend = shc.dendrogram(shc.linkage(X_adolescents, method='ward'))
plt.show()

# Определение оптимального количества кластеров методом локтя для детей
inertia_children = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_children)
    inertia_children.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), inertia_children, marker='o')
plt.title("Метод локтя для детей")
plt.xlabel("Количество кластеров")
plt.ylabel("Инерция")
plt.show()

# Определение оптимального количества кластеров методом силуэта для детей
silhouette_scores_children = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_children)
    score = silhouette_score(X_children, kmeans.labels_)
    silhouette_scores_children.append(score)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), silhouette_scores_children, marker='o')
plt.title("Метод силуэта для детей")
plt.xlabel("Количество кластеров")
plt.ylabel("Коэффициент силуэта")
plt.show()

# Оптимальное количество кластеров для детей
optimal_clusters_children = 3

# Определение оптимального количества кластеров методом локтя для подростков
inertia_adolescents = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_adolescents)
    inertia_adolescents.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), inertia_adolescents, marker='o')
plt.title("Метод локтя для подростков")
plt.xlabel("Количество кластеров")
plt.ylabel("Инерция")
plt.show()

# Определение оптимального количества кластеров методом силуэта для подростков
silhouette_scores_adolescents = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X_adolescents)
    score = silhouette_score(X_adolescents, kmeans.labels_)
    silhouette_scores_adolescents.append(score)

plt.figure(figsize=(8, 4))
plt.plot(range(2, 10), silhouette_scores_adolescents, marker='o')
plt.title("Метод силуэта для подростков")
plt.xlabel("Количество кластеров")
plt.ylabel("Коэффициент силуэта")
plt.show()

# Оптимальное количество кластеров для подростков
optimal_clusters_adolescents = 3

# Иерархическая кластеризация для детей
agglo_children = AgglomerativeClustering(n_clusters=optimal_clusters_children, affinity='euclidean', linkage='complete')
children_clusters = agglo_children.fit_predict(X_children)

# Иерархическая кластеризация для подростков
agglo_adolescents = AgglomerativeClustering(n_clusters=optimal_clusters_adolescents, affinity='euclidean', linkage='complete')
adolescents_clusters = agglo_adolescents.fit_predict(X_adolescents)

# Визуализация кластеров для детей в 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_children[:, 0], X_children[:, 1], X_children[:, 2], c=children_clusters, cmap='viridis')
ax.set_title("Кластеры для детей")
ax.set_xlabel('Серия 1')
ax.set_ylabel('Серия 2')
ax.set_zlabel('Серия 3')
plt.show()

# Визуализация кластеров для подростков в 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_adolescents[:, 0], X_adolescents[:, 1], X_adolescents[:, 2], c=adolescents_clusters, cmap='viridis')
ax.set_title("Кластеры для подростков")
ax.set_xlabel('Серия 1')
ax.set_ylabel('Серия 2')
ax.set_zlabel('Серия 3')
plt.show()
```

```python
# Подключение к Google Drive
from google.colab import drive
drive.mount('/content/gdrive')

# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from mpl_toolkits.mplot3d import Axes3D

# Загрузка данных
file_path = "/content/gdrive/My Drive/working_memory/child_adolesc.csv"
data = pd.read_csv(file_path)

# Разделение данных на две группы: дети (0) и подростки (1)
children_data = data[data['Ребенок(0)/Подросток(1)'] == 0]
adolescents_data = data[data['Ребенок(0)/Подросток(1)'] == 1]

# Извлечение данных для анализа
X_children = children_data[['WMem: Количество объектов, серия 1',
                            'WMem: Количество объектов, серия 2',
                            'WMem: Количество объектов, серия 3']].values

X_adolescents = adolescents_data[['WMem: Количество объектов, серия 1',
                                  'WMem: Количество объектов, серия 2',
                                  'WMem: Количество объектов, серия 3']].values

# Функция для построения дендрограммы (только для метода Ward с евклидовой метрикой)
def plot_dendrogram(X, title):
    plt.figure(figsize=(10, 7))
    plt.title(f"Дендрограмма (Ward - Euclidean) - {title}")
    dend = shc.dendrogram(shc.linkage(X, method='ward'))
    plt.show()

# Функция для выполнения кластеризации и визуализации в 3D с различными метриками
def plot_clusters(X, title, affinity_metric, linkage_method):
    # Определение количества кластеров (выбор числа кластеров зависит от дендрограммы)
    optimal_clusters = 4  # Выберите количество кластеров на основе дендрограммы

    # Иерархическая кластеризация
    agglo = AgglomerativeClustering(n_clusters=optimal_clusters, affinity=affinity_metric, linkage=linkage_method)
    clusters = agglo.fit_predict(X)

    # Визуализация кластеров в 3D
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=clusters, cmap='viridis')
    ax.set_title(f"Кластеры ({affinity_metric}) - {title}")
    ax.set_xlabel('Серия 1')
    ax.set_ylabel('Серия 2')
    ax.set_zlabel('Серия 3')
    plt.show()

# Построение дендрограммы для детей и подростков
plot_dendrogram(X_children, "Дети")
plot_dendrogram(X_adolescents, "Подростки")

# Проверка различных метрик на данных для детей
for metric in ['euclidean', 'manhattan', 'cosine', 'chebyshev']:
    plot_clusters(X_children, "Дети", metric, 'average')  # Используем метод 'average'

# Проверка различных метрик на данных для подростков
for metric in ['euclidean', 'manhattan', 'cosine', 'chebyshev']:
    plot_clusters(X_adolescents, "Подростки", metric, 'average')  # Используем метод 'average'
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Нормализация данных
scaler = StandardScaler()
X_children_scaled = scaler.fit_transform(X_children)
X_adolescents_scaled = scaler.fit_transform(X_adolescents)

# Применение PCA для понижения размерности (если данных много)
pca = PCA(n_components=3)  # Оставляем 3 компоненты для 3D визуализации
X_children_pca = pca.fit_transform(X_children_scaled)
X_adolescents_pca = pca.fit_transform(X_adolescents_scaled)

# Применение различных методов связи с изменением метрики расстояния
for linkage_method in ['single', 'complete', 'average']:
    for metric in ['euclidean', 'manhattan']:
        plot_clusters(X_children_pca, f"Дети - {linkage_method}", metric, linkage_method)
        plot_clusters(X_adolescents_pca, f"Подростки - {linkage_method}", metric, linkage_method)
```
